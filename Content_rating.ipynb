```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('/content/content  rating with string.csv')
X = dataset.iloc[:, :1].values
```

```python
print (X)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
print(X)
```

```python
 pd.DataFrame(X.tolist(), columns=['content_rating'])
```

```python
 pd.DataFrame(X.tolist(), columns=['content_rating']).to_csv("Vectors for content_rating.csv")
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(X.tolist())
```

```python
print(similarity)
```

```python
pd.DataFrame(similarity).to_csv("Similarity_content_rating.csv")
```
